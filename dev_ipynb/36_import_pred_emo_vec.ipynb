```python
import pandas as pd
import torch
```

```python
csv_file = 'emo_pred_list_example.csv'
df = pd.read_csv(csv_file)
df
```

```python
#emotions = ["sad", "happy", "fearful", "neutral", "disgusted", "angry", "surprised"]
convert_emo_str = {
    'happiness':'happy',
    'disgust':'disgusted',
    'fear':'fearful',
    'sadness':'sad',
    'surprise':'surprised'
}
df = df.rename(columns=convert_emo_str)
```

```python
df['file_name'] = df.file_name.apply(convert_file_name)
```
