# Hard clipping function

```python
import matplotlib.pyplot as plt
import numpy as np
import torch
```

```python
import torch.nn.functional as F
def hard_clip(input):
    return F.hardtanh(input + 0.5 , min_val=0, max_val=1)
```

```python
x = torch.Tensor(np.arange(-5,5,0.01))
y = hard_clip(x)
```

```python
plt.plot(x, y)
plt.show()
```

```python
y2 = F.sigmoid(x)
plt.plot(x, y)
plt.plot(x, y2)
plt.show()
```

# Soft clipping function

$$SC_{p}(x) = \frac{1}{p} \log \left ( \frac{1 + e^{px}}{1 + e^{p(x-1)}} \right )$$

```python
import torch
import torch.nn.functional as F
import numpy as np
```

```python
def soft_clip(input, p=50):
    return torch.log((1 + torch.exp(p * x)) / (1 + torch.exp(p * (x - 1)))) / p
```

```python
def soft_clip(input, p=50):
    return torch.log(1 + torch.exp(p * x)) / p - torch.log(1 + torch.exp(p * (x - 1))) / p
```

```python
x = torch.Tensor(np.arange(-5,5,0.01))
```

```python
p = 50
y = soft_clip(x, p)

plt.plot(x, y)
plt.show()
```
