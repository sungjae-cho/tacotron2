# Exploring the YIN pitch detection algorithm

```python
import librosa
import IPython.display as ipd
from yin import compute_yin
import matplotlib.pyplot as plt
```

```python
wav_path = '../../data_tts/BC2013/wav/CA-BB-01-02.wav'
sig, sr = librosa.load(wav_path)
```

```python
ipd.display(ipd.Audio(wav_path, rate=sr))
```

```python
plt.plot(sig)
```

```python
pitches, harmonic_rates, argmins, times = compute_yin(
    sig, sr, w_len=512, w_step=256, f0_min=70, f0_max=500,
    harmo_thresh=0.1)
plt.ylim(0, 500)
plt.plot(pitches)
```
