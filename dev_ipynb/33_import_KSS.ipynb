```python
import os
cwd_old = os.getcwd()
dir_path = '/data2/sungjaecho/data_tts/KSS/KSS'
os.chdir(dir_path)
cwd_new = os.getcwd()
print("The working directory is moved from {} to {}.".format(cwd_old, cwd_new))
```

```python
from glob import glob
import pandas as pd
import random
```

```python
old_wavpaths = sorted(glob(cwd_new + '/*/*.wav', recursive=True))
```

```python
len(old_wavpaths)
```

```python
with open('transcript.v.1.4.txt') as f:
    transcript = f.read()
lines = transcript.split('\n')[:-1] # The last one is the empty string.
print(len(lines))
```

```python
i = 0 
for line in lines:
    n_elements = len(line.split('|'))
    if n_elements != 6:
        print(n_elements)
        print(line)
        print(i)
    i += 1
```

```python
str_speaker = 'kss-w'
wavpaths = list()
durations = list()
text_sentences = list()
for line in lines:
    elements = line.split('|')
    wavpath = elements[0]
    sentence = elements[2]
    duration = elements[4]
    
    _, wav_name = os.path.split(wavpath)
    wav_name = wav_name.replace('_', '-')
    new_wavpath = "{}/wav/{}_{}".format(dir_path, str_speaker, wav_name)
    
    wavpaths.append(new_wavpath)
    durations.append(duration)
    text_sentences.append(sentence)
```

```python
len(old_wavpaths) * 0.01
```

```python
splits = ['val'] * 128 + ['test'] * 128 + ['train'] * (len(wavpaths) - 2*128)
random.seed(1234)
random.shuffle(splits)
```

```python
len(splits) == len(wavpaths)
```

```python
df = pd.DataFrame({
    'database':['kss']*len(wavpaths),
    'split':splits,
    'wav_path':wavpaths,
    'duration':durations,
    'text':text_sentences,
    'speaker':[str_speaker]*len(wavpaths),
    'sex':['w']*len(wavpaths),
    'emotion':['neutral']*len(wavpaths),
    'lang':['ko']*len(wavpaths),
})
df
```
