# Throw out long wav files

```python
import os
import pandas as pd
import random
import IPython.display as ipd
```

# 1. KETTS

```python
db_csv_path = os.path.join('metadata', 'ketts.csv'); print(db_csv_path)
```

```python
df_ketts = pd.read_csv(db_csv_path)
df_ketts.head()
```

```python
df_ketts.duration.describe()
```

```python
df_ketts.duration.quantile(0.8)
```

```python
wav_path = df_ketts[(df_ketts.duration > 40)].wav_path.to_numpy()[0]
print(wav_path)
ipd.Audio(wav_path)
```

```python
print("Ratio of samples longer than 10 secnonds:")
df_ketts[(df_ketts.duration > 10)].shape[0] / df_ketts.shape[0] 
```

```python
print("Ratio of duraion of samples longer than 10 secnonds:")
df_ketts[(df_ketts.duration > 10)].duration.sum() / df_ketts.duration.sum()
```

**Result**:
- 8.7% of samples are longer than 10 seoncds.

**Conclusion**
- It is okay to delete samples long than 10 seconds.
